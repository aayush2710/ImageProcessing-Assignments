```python
import numpy as np
import scipy.stats
import sklearn.datasets
```

```python
N = 100
D = 2
K = 5
```

```python
U0 = np.random.random((K,D))
V0 = np.zeros((K,D,D))
for i in range(K):
        V0[i] = sklearn.datasets.make_spd_matrix(D)
Z0 = np.random.random((K,))
Z0 = Z0/np.sum(Z0)
```

```python
V0.shape
```

```python
X = np.random.random((N,D))
```

```python
Rnk = np.zeros((N,K))
Uk = U0
Ek = V0
PIk = Z0
Prev = -np.inf
epsilon = 1e-3
for _ in range(250):
    for i in range(N):
        Sum = 0
        for j in range(K):
            Rnk[i,j] = scipy.stats.multivariate_normal.pdf(X[i],Uk[j], Ek[j]) * PIk[j]
            Sum += Rnk[i,j]
        Rnk[i]=Rnk[i]/Sum
        
    Nk = np.sum(Rnk , axis = 0).reshape(-1,1)
    
    Uk = np.zeros((K,D))
    for j in range(K):
        Uk[j] = np.sum(Rnk[:,j:j+1]*X , axis = 0)
    Uk=Uk/Nk

    Ek = np.zeros((K,D,D))
    for i in range(N):
        for j in range(K):
            Ek[j] += Rnk[i,j:j+1]*(np.dot((X[i]-Uk[j]).reshape(D,1) , (X[i]-Uk[j]).reshape(1,D) ))
    Ek/=Nk.reshape(-1,1,1)

    PIk = Nk/N
    
    
    L = 0
        
    for i in range(N):
            Sum = 0
            for j in range(K):
                Sum += PIk[j]*scipy.stats.multivariate_normal.pdf(X[i],Uk[j], Ek[j])
            L += np.log(Sum)
        
    print("Log likelihood : ", L)
    if(np.abs(L - Prev) < epsilon):
        break
    else:
        Prev = L
    print("Mu_k :" , Uk)
    print("E_k :" , Ek)
    print("Pi_k :" , PIk)
        
```
